# Jupyter Kernel Control and Debugging Milestone
This lesson will help you master running, restarting, and interrupting Jupyter kernels. Understanding kernel control is essential for debugging, reproducibility, and maintaining a clean notebook state.

## 1. What is a Jupyter Kernel?
A Jupyter kernel is the computational engine that executes the code in your notebook. It remembers variables, imports, and state until you restart it. Managing the kernel is crucial for debugging and reproducibility.

## 2. Running Cells and Understanding Execution Order
Run the following cells one by one. Notice how the output depends on the order of execution.

```python
# Cell 1: Define a variable
x = 10
```

```python
# Cell 2: Use the variable
print(f"The value of x is: {x}")
```

- Try running Cell 2 before Cell 1. What happens?
- Now run Cell 1, then Cell 2. Notice the difference.

## 3. Restarting the Kernel
Restarting the kernel clears all variables and resets the notebook state.

- Use the Jupyter menu to restart the kernel now.
- After restarting, try running only Cell 2. You should get an error because `x` is no longer defined.
- Rerun Cell 1, then Cell 2 to restore the state.

## 4. Interrupting Execution
Sometimes, code runs too long or gets stuck. You can interrupt the kernel to stop execution without restarting.

```python
# Cell 3: Deliberately long-running operation
import time
for i in range(100):
    print(f"Loop {i}")
    time.sleep(1)  # Wait 1 second per loop
```

- Run the cell above. While it is running, use the Jupyter menu to **Interrupt** the kernel.
- Confirm that execution stops and the notebook remains responsive.

## 5. When to Restart vs. Interrupt
- **Interrupt** is useful for stopping code that is running too long or is stuck, but keeps your variables and imports intact.
- **Restart** is necessary when the kernel is unresponsive, or when you want to clear all variables and start fresh.
### Scenarios:
- If a cell is taking too long, try **Interrupt** first.
- If you see strange errors or variables have unexpected values, use **Restart** and rerun all cells in order.
- If the kernel is completely frozen, use **Restart**.
