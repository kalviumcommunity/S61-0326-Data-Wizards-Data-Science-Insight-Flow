## Section 1: Working with Numeric Data Types

Python has two main numeric types:
- **Integers (int)**: Whole numbers like 42, -10, 0
- **Floats (float)**: Decimal numbers like 3.14, -0.5

Let's explore them!

```python
# 1.1 Working with Integers and Floats

age = 25
temperature = -10
pi = 3.14159
balance = -25.50

print("Numeric Types Examples:")
print(f"  age = {age} (type: {type(age).__name__})")
print(f"  temperature = {temperature} (type: {type(temperature).__name__})")
print(f"  pi = {pi} (type: {type(pi).__name__})")
print(f"  balance = {balance} (type: {type(balance).__name__})")

print("\nArithmetic with Integers:")
print(f"  {age} + 5 = {age + 5}")
print(f"  {age} * 2 = {age * 2}")
print(f"  {age} // 3 = {age // 3}  (floor division)")
print(f"  {age} % 3 = {age % 3}    (remainder)")

print("\nDivision: / vs //")
print(f"  10 / 4 = {10 / 4}   (type: {type(10 / 4).__name__})")
print(f"  10 // 4 = {10 // 4}   (type: {type(10 // 4).__name__})")
print("  → / always returns float, // returns integer")
```

## Section 2: Understanding String Data Types

Strings represent text in Python. They must be enclosed in quotes (' or ").

```python
# 2.1 Creating and Working with Strings

name = "Alice"
city = "New York"

print("String Examples:")
print(f"  name = '{name}' (type: {type(name).__name__})")
print(f"  len(name) = {len(name)}")

print("\nIndexing (accessing characters):")
word = "Python"
print(f"  word[0] = '{word[0]}'    (first character)")
print(f"  word[-1] = '{word[-1]}'   (last character)")

print("\nSlicing (accessing groups):")
sentence = "Data Science"
print(f"  '{sentence}'[0:4] = '{sentence[0:4]}'")
print(f"  '{sentence}'[5:] = '{sentence[5:]}'")
print(f"  reversed: '{sentence[::-1]}'")

print("\nString Methods:")
text = "Hello World"
print(f"  '{text}'.lower() = '{text.lower()}'")
print(f"  '{text}'.upper() = '{text.upper()}'")
print(f"  '{text}'.replace('World', 'Python') = '{text.replace('World', 'Python')}'")
print(f"  '{text}'.split() = {text.split()}")

print("\nString Concatenation and Formatting:")
first = "John"
last = "Doe"
age = 30
message = f"{first} {last} is {age} years old"
print(f"  Using f-string: '{message}'")
```

## Section 3: Mixing Numbers and Strings Safely

### The Core Challenge
- `42` is a number you can do math with
- `"42"` is text that *looks* like a number
- Mixing them incorrectly causes errors!

You need to **explicitly convert** between types.

```python
# 3.1 Type Conversion

age_number = 25
age_string = "25"

print("Type Comparison:")
print(f"  age_number = {age_number} (type: {type(age_number).__name__})")
print(f"  age_string = '{age_string}' (type: {type(age_string).__name__})")

print("\n✓ This works:")
print(f"  {age_number} + 5 = {age_number + 5}")

print("\n✗ This FAILS:")
try:
    result = age_string + 5
except TypeError as e:
    print(f"  '{age_string}' + 5 → TypeError: {e}")

print("\n" + "="*60)

# 3.2 Converting String to Number
print("\nConverting String → Number:")
string_number = "42"
string_decimal = "3.14"

num_int = int(string_number)
num_float = float(string_decimal)

print(f"  int('{string_number}') = {num_int}")
print(f"  float('{string_decimal}') = {num_float}")
print(f"  Now math works: {num_int} + 8 = {num_int + 8}")

# 3.3 Converting Number to String
print("\nConverting Number → String:")
age = 30
price = 19.99

age_str = str(age)
price_str = str(price)

print(f"  str({age}) = '{age_str}'")
print(f"  str({price}) = '{price_str}'")
print(f"  Now concatenation works: 'Age: ' + '{age_str}' = 'Age: {age_str}'")
```

## Section 4: Inspecting Data Types

### Why Check Types?
- Prevent errors before they happen
- Debug problems quickly
- Write defensive, reliable code
- Build good programming habits

```python
# 4.1 Using type() Function

print("Checking Data Types:\n")

variables = {
    "integer": 42,
    "float": 3.14,
    "string": "Hello",
    "string_number": "42",
}

for name, value in variables.items():
    data_type = type(value).__name__
    print(f"  type({name:15}) = {data_type:10} (value: {value!r})")

print("\n" + "="*60)

# 4.2 Using isinstance() Function

print("\nUsing isinstance() for Type Checking:\n")

value = 25
print(f"value = {value}\n")
print(f"  isinstance(value, int) = {isinstance(value, int)}")
print(f"  isinstance(value, float) = {isinstance(value, float)}")
print(f"  isinstance(value, str) = {isinstance(value, str)}")
print(f"  isinstance(value, (int, float)) = {isinstance(value, (int, float))}")
print("  (Last one: Is it EITHER int OR float?)")

print("\n" + "="*60)

# 4.3 Type-Safe Function

print("\nType-Safe Function Example:\n")

def process_value(val):
    """Safe processing based on type."""
    if isinstance(val, (int, float)):
        return val * 2
    elif isinstance(val, str):
        try:
            return float(val) * 2
        except ValueError:
            return f"Cannot convert '{val}' to number"
    else:
        return f"Unknown type: {type(val).__name__}"

tests = [10, 5.5, "20", "3.14", "invalid", [1, 2, 3]]
print("Testing with different inputs:")
for t in tests:
    print(f"  process_value({t!r:15}) = {process_value(t)}")
```

## Final Example: Temperature Data Processing

### Scenario
You have raw temperature data with mixed types. Your job:
1. Identify types
2. Convert to consistent type
3. Perform calculations
4. Generate report

```python
# Temperature Data Processing Example

raw_data = {
    "Monday": 22,
    "Tuesday": "23.5",
    "Wednesday": 19,
    "Thursday": "25.8",
    "Friday": 21,
}

print("TEMPERATURE DATA PROCESSING\n")
print("Step 1: Identify Types")
for day, temp in raw_data.items():
    print(f"  {day}: {temp!r} (type: {type(temp).__name__})")

print("\nStep 2: Convert All to Float")
temps_processed = {}
for day, temp in raw_data.items():
    temp_float = float(temp)
    temps_processed[day] = temp_float
    print(f"  {day}: {temp!r} → {temp_float}")

print("\nStep 3: Calculate Statistics")
temps = list(temps_processed.values())
avg = sum(temps) / len(temps)
max_temp = max(temps)
min_temp = min(temps)

print(f"  Average: {avg:.2f}°C")
print(f"  Maximum: {max_temp:.1f}°C")
print(f"  Minimum: {min_temp:.1f}°C")

print("\nStep 4: Generate Report")
report = f"""
WEEKLY TEMPERATURE REPORT
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Average: {avg:.2f}°C
Maximum: {max_temp:.1f}°C  
Minimum: {min_temp:.1f}°C
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

This week's average temperature was {avg:.2f}°C,
with a high of {max_temp:.1f}°C and a low of {min_temp:.1f}°C.
"""

print(report)
```
