# Understanding Notebook Cells: Code vs Markdown

**Purpose:** This notebook demonstrates the correct use of Code cells and Markdown cells in Jupyter Notebooks.

- **Code cells** are used to write and execute Python logic
- **Markdown cells** are used to explain, describe, and provide context

By separating execution from explanation, notebooks become easier to read, review, and collaborate on.

## Section 1: Working with Variables

The code cell below defines a few simple Python variables and prints their values.

This is an example of what belongs in a **Code cell** — pure execution logic with no narrative mixed in.

```python
# Defining simple variables
name = "Data Science Student"
year = 2025
is_learning = True

print(f"Name    : {name}")
print(f"Year    : {year}")
print(f"Learning: {is_learning}")
```

### What the output tells us

The code above assigned three variables — a string, an integer, and a boolean — and printed them using an f-string.

This is an example of what belongs in a **Markdown cell** — interpreting the output and explaining its significance, not re-executing any logic.

## Section 2: Basic Arithmetic Operations

The next code cell performs a few simple arithmetic calculations.

Even simple operations benefit from a Markdown explanation before them — it tells the reader *why* you are doing the calculation, not just *what* the code does.

```python
# Basic arithmetic
a = 15
b = 4

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b:.2f}")
print(f"{a} // {b} = {a // b}  (floor division)")
print(f"{a} % {b}  = {a % b}   (remainder)")
```

### What the output tells us

The cell above demonstrated six arithmetic operators in Python:

- `+`, `-`, `*`, `/` — standard operations
- `//` — floor division, which drops the decimal and returns the whole number
- `%` — the modulo operator, which returns the remainder after division

Writing this interpretation in a **Markdown cell** (rather than as code comments) makes the notebook readable even to someone who is not a Python programmer.

## Summary: Code vs Markdown — When to Use Each

| Cell Type | Use It For | Do NOT Use It For |
|-----------|-----------|-------------------|
| **Code** | Running Python, computations, logic | Writing explanations or narrative |
| **Markdown** | Titles, context, reasoning, interpreting output | Executing any code |

### Why this separation matters

- **Readability:** Reviewers can skim Markdown headings to understand the notebook's flow without reading every line of code
- **Collaboration:** Teammates understand *why* decisions were made, not just *what* was executed
- **Professionalism:** Well-structured notebooks are a core expectation in real Data Science workflows

> A notebook that only uses code comments is a script, not a document. Markdown cells are what make a notebook a story.
